```python
merged_df = pd.merge(netflix_dataset, df_tittle[['Movie_Id', 'Name']], on='Movie_Id', how='left')
best_movies = merged_df.groupby(['Movie_Id', 'Name'])['Rating'].mean().sort_values(ascending=False)
print(best_movies)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the top 10 most liked movies from the best_movies series
top_10_liked_movies = best_movies.head(10).reset_index()

plt.figure(figsize=(13, 9))
sns.barplot(x='Rating', y='Name', data=top_10_liked_movies, palette='viridis')
plt.title('Top 10 Most Liked Movies by Average Rating', fontsize=16)
plt.xlabel('Average Rating', fontsize=16)
plt.ylabel('Movie Name', fontsize=16)
plt.xlim(top_10_liked_movies['Rating'].min() * 0.9, 5.0) # Set x-axis limit for better visualization
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
